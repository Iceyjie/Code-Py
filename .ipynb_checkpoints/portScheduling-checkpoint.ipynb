# Port Scheduling Problem under Uncertain Arrival time

```python
import xml.etree.ElementTree as ET
import pandas as pd
import os
import zipfile
from collections import deque
import matplotlib.pyplot as plt
```

```python
def createFolder(folder_path):
    os.makedirs(folder_path, exist_ok=True)
```

# Data Preparation and Visiualization

```python
def extractZip(xml_path, zip_path):
    zip_set = [zip for zip in os.listdir(zip_path) if (zip != '.DS_Store') and (zip.endswith(".zip"))]
    for zip in zip_set:
        xml_folder_name = os.path.splitext(os.path.basename(zip))[0]
        extract_folder = os.path.join(xml_path, xml_folder_name)
        createFolder(extract_folder)
        with zipfile.ZipFile(os.path.join(zip_path, zip), 'r') as zip_data:
            for member in zip_data.namelist():
                filename = os.path.basename(member)
                if not filename:
                    continue  
                source = zip_data.open(member)
                target_path = os.path.join(extract_folder, filename)
                with open(target_path, "wb") as target:
                    target.write(source.read())
```

```python
def getCSV(csv_path, merge_path, xml_path, start_name, data_type):
    xml_folders = [folder for folder in os.listdir(xml_path) if (folder != '.DS_Store') and (folder.startswith(start_name))] #all XML folder names
    error_files = []
    xml_folder_data = pd.DataFrame()
    for xml_folder in xml_folders: 

        # print(f'xml_file_name is {xml_folder}')

        xml_files = [f for f in os.listdir(os.path.join(xml_path, xml_folder)) if f.endswith(".XML")]
        csv_folder = os.path.join(csv_path, xml_folder)
        createFolder(csv_folder)
        for xml_file in xml_files:

            # print(f'processing xml_file_name is {xml_file}')

            csv_file = xml_file.replace('.XML', '.csv')
            try:
                tree = ET.parse(os.path.join(xml_path, xml_folder, xml_file))
                root = tree.getroot()
                data = []
                for entry in root.findall('G_SQL1'):
                    row = {child.tag: child.text for child in entry}
                    data.append(row)
                df = pd.DataFrame(data)
                df.to_csv(os.path.join(csv_folder, csv_file), index=False)
                xml_folder_data = pd.concat([xml_folder_data, df], ignore_index=True)
            except ET.ParseError as e:
                print(f"skip destroyed file: {xml_file}")
                error_files.append((xml_file, str(e)))
                continue
    xml_folder_data.to_csv(os.path.join(merge_path, data_type[start_name] + '.csv'), index=False)
```

```python
def mergeTime(csv_path, merge_filename, min_delay, max_delay):
    data_name_set = ['ARRIVED', 'DEPARTED', 'IN_PORT', 'DUE']
    file_paths = {file:os.path.join(csv_path, file+'.csv') for file in data_name_set}

    dataframes = {
        name: pd.read_csv(path).rename(columns=lambda c: c.strip().upper())
        for name, path in file_paths.items()
    }
    vessel_names = dataframes['ARRIVED']['VESSEL_NAME'].dropna().unique()

    merged_rows = []
    for vessel in vessel_names:
        row = {'VESSEL_NAME': vessel}
        for name, df in dataframes.items():
            match = df[df['VESSEL_NAME'] == vessel]
            if not match.empty:
                for col in match.columns:
                    if col != 'VESSEL_NAME':
                        row[f'{name}_{col}'] = match.iloc[0][col]
        merged_rows.append(row)
    
    merged_df = pd.DataFrame(merged_rows)
    merged_df['DUE_ETA'] = pd.to_datetime(merged_df['DUE_ETA'], format="%Y/%m/%d %H:%M", errors='coerce')
    merged_df['ARRIVED_ARRIVAL_TIME'] = pd.to_datetime(merged_df['ARRIVED_ARRIVAL_TIME'], format="%d-%b-%Y %H:%M", errors='coerce')
    merged_df['ArrivalDelay'] = (merged_df['ARRIVED_ARRIVAL_TIME'] - merged_df['DUE_ETA']).dt.total_seconds() / 3600
    merged_df = merged_df[(merged_df["ArrivalDelay"] >= min_delay) & (merged_df["ArrivalDelay"] <= max_delay)]
    merged_df.to_csv(os.path.join(csv_path, merge_filename), index=False)
```

```python
#parameter
xml_path = '../Data/XML/'
csv_path = '../Data/CSV/'
zip_path = '../Data/ZIP/'
merge_path = '../Data/MERGE/'
data_type = {'c9a': 'ARRIVED', '7cd': 'DUE', '689': 'IN_PORT', '024': 'DEPARTED'}
merge_filename = 'mergeResult.csv' 
min_delay = -24
max_delay = 24

#process data
createFolder(merge_path)
extractZip(xml_path=xml_path, zip_path=zip_path)
for start_name in data_type.keys():
    getCSV(csv_path=csv_path, merge_path=merge_path,xml_path=xml_path, start_name=start_name, data_type=data_type)
mergeTime(csv_path=merge_path, merge_filename=merge_filename, min_delay=min_delay, max_delay=max_delay)
print('--- merge data successfully ---')
```

```python
merge_filename = 'mergeResult.csv' 
data = pd.read_csv(os.path.join(merge_path, merge_filename))
plt.figure()
plt.hist(data['ArrivalDelay'].dropna(), bins=50, color='skyblue', edgecolor='black')
plt.title("ATA minus ETA (Hour)")
plt.xlabel("Arrival Delay (Hour)")
plt.ylabel("Count")
plt.show()
```
